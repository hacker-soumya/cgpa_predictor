```python
import pandas as pd
```

```python
df=pd.read_csv("Student_Performance_2026.csv")
```

```python
df.describe()
```

```python
df=df[df["Age"]>18]
```

```python
df=df[df["Age"]<24]
```

```python
df.drop(columns="student_ID",inplace=True)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,GridSearchCV 
from sklearn.metrics import accuracy_score,mean_absolute_error, mean_squared_error, r2_score
```

```python
df
```

```python
df1=pd.get_dummies(df,columns=["Gender"],drop_first=True,dtype=int)
```

```python
df1
```

```python
X = df1.drop(columns=["current_sem_CGPA"])
y = df1["current_sem_CGPA"]
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
parameters = {
    "max_depth": [1,2,3,5,7,None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "criterion": ["squared_error", "friedman_mse", "absolute_error"],
   
}
```

```python
model=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=parameters,cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
best=model.best_estimator_
```

```python
print(model.best_params_)
```

```python
y_pred=best.predict(X_test)
```

```python
print(y.name)
print(X.columns.tolist())
print(df.shape)
print(y.name in X.columns)
corr = df.corr(numeric_only=True)
print(corr[y.name].sort_values(ascending=False))
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
print("R² Score:", best.score(X_train, y_train))
print("R² Score:", best.score(X_test, y_test))
```
